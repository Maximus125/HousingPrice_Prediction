```python
df = df.drop(['id'], axis=1)
```

```python
df.head()
```

### Formatting the date

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
df['year'] = df['date'].apply(lambda yr: yr.year)
df.head()
```

```python
df.groupby('year').mean()['price'].plot(figsize=(10,5))
```

```python
df.plot(kind='scatter', x='long', y='lat',s=df["price"]/100000, label="price", c='price', cmap=plt.get_cmap('RdBu'), colorbar=True, alpha=0.3, figsize=(10,8))
```

```python
df['renovated'] = df['yr_renovated'].apply(lambda x: 0 if x==0 else 1)
```

```python
df.head()
```

```python
df['renovated'].value_counts()
```

```python
df.groupby('zipcode')['price'].mean().sort_values(ascending=False).head(10)
```

### Dropping some columns which aren't important

```python
df=df.drop(['date','year','zipcode','yr_renovated'], axis=1)
```
